<h3> Testes dos métodos implementados em C

```python
import ctypes
import numpy as np
import matplotlib.pyplot as plt
# Carrega a biblioteca
lib = ctypes.CDLL("./metodos.dll")
```

```python
epsilon = 0.000001
FUNC_TYPE = ctypes.CFUNCTYPE(ctypes.c_double, ctypes.c_double)
# Metódo da Bisseção
lib.metodo_bissecao.restype = ctypes.c_double #Retorno
lib.metodo_bissecao.argtypes = [FUNC_TYPE,ctypes.c_double,ctypes.c_double,ctypes.c_double]

#Funções de teste
lib.f1.restype = ctypes.c_double #retorno
lib.f1.argtypes = [ctypes.c_double] #argumentos
f1_cfunc = FUNC_TYPE(lib.f1)

lib.f2.restype = ctypes.c_double #retorno
lib.f2.argtypes = [ctypes.c_double] #argumentos
f2_cfunc = FUNC_TYPE(lib.f2)

lib.f3.restype = ctypes.c_double #retorno
lib.f3.argtypes = [ctypes.c_double] #argumentos
f3_cfunc = FUNC_TYPE(lib.f3)

lib.f4.restype = ctypes.c_double #retorno
lib.f4.argtypes = [ctypes.c_double] #argumentos
f4_cfunc = FUNC_TYPE(lib.f4)
```

<h2>Método da Bisseção

<h4>Teste 1 f(x)= x² - 2

```python
# Eixos
ax = plt.gca()
ax.spines['bottom'].set_position('zero')
ax.spines['left'].set_position('zero')

# Calcula usando C
xs = np.linspace(-3, 3, 100)
ys = [lib.f1(float(x)) for x in xs]

# Plota o gráfico da função
plt.plot(xs, ys,color = 'black')

xz = lib.metodo_bissecao(f1_cfunc,1,3, epsilon)
print(xz)
plt.scatter(xz, 0, color = 'red')

plt.show()
```

<h2>Newton Raphson

```python
FUNC_TYPE = ctypes.CFUNCTYPE(ctypes.c_double, ctypes.c_double)
# Newton Raphson
lib.newton_raphson.restype = ctypes.c_double #Retorno
lib.newton_raphson.argtypes = [FUNC_TYPE, FUNC_TYPE, ctypes.c_double, ctypes.c_double]

#Funções de teste
lib.f1.restype = ctypes.c_double #retorno
lib.f1.argtypes = [ctypes.c_double] #argumentos
f1_cfunc = FUNC_TYPE(lib.f1)

lib.f2.restype = ctypes.c_double #retorno
lib.f2.argtypes = [ctypes.c_double] #argumentos
f2_cfunc = FUNC_TYPE(lib.f2)

lib.f3.restype = ctypes.c_double #retorno
lib.f3.argtypes = [ctypes.c_double] #argumentos
f3_cfunc = FUNC_TYPE(lib.f3)

lib.f4.restype = ctypes.c_double #retorno
lib.f4.argtypes = [ctypes.c_double] #argumentos
f4_cfunc = FUNC_TYPE(lib.f4)

lib.d1.restype = ctypes.c_double #retorno
lib.d1.argtypes = [ctypes.c_double] #argumentos
d1_cfunc = FUNC_TYPE(lib.d1)

lib.d2.restype = ctypes.c_double #retorno
lib.d2.argtypes = [ctypes.c_double] #argumentos
d2_cfunc = FUNC_TYPE(lib.d2)

lib.d3.restype = ctypes.c_double #retorno
lib.d3.argtypes = [ctypes.c_double] #argumentos
d3_cfunc = FUNC_TYPE(lib.d3)

lib.d4.restype = ctypes.c_double #retorno
lib.d4.argtypes = [ctypes.c_double] #argumentos
d4_cfunc = FUNC_TYPE(lib.d4)
```
